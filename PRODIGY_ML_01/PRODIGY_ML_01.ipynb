```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# read train.csv
df = pd.read_csv("train.csv")
df.head()
```

```python
# defining the input varables and target variable
# here we take features which consist of area , bedroom and bathroom to take whole area of house and get total count of bedrooms and bathrooms.
# for target, we need only SalePrice
x = df[["LotArea","TotalBsmtSF","1stFlrSF","2ndFlrSF","GrLivArea","FullBath","HalfBath","BedroomAbvGr"]]
y = df.SalePrice
```

```python
# split the data into train and test data that use to find accuracy of a model
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# train a model using training data
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# test a model using test data
model.score(x_test,y_test)
```

```python
# now we predict the price of test.csv file using this model
df2 = pd.read_csv("test.csv")
df2.head()
```

```python
# extract required field from test.csv and fill missing data
x_new = df2[["LotArea","TotalBsmtSF","1stFlrSF","2ndFlrSF","GrLivArea","FullBath","HalfBath","BedroomAbvGr"]]
x_new.fillna(0,inplace=True)
```

```python
y_new = model.predict(x_new)
```

```python
df2["price"] = y_new
```
